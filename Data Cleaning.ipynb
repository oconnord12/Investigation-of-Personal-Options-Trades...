```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Data/Apr 16, 2020 – Jun 8, 2023.csv', on_bad_lines='skip')
df
```

```python
df.drop(columns=['Process Date', 'Settle Date'],inplace=True)
```

```python
df['Trans Code'].unique()
```

```python
options_codes = ['BTO','STC','OEXP'] #buy to order, sell to close, option expired, respectively. 
simple_options_df = df.copy()
simple_options_df = simple_options_df[simple_options_df['Trans Code'].isin(options_codes)]
simple_options_df.sample(10)
```

```python
simple_options_df['Trade Type'] = np.nan
simple_options_df['Trade Type'] = simple_options_df['Description'].apply(lambda x: 'Call' if 'Call' in x else 'Put')
simple_options_df.sample(10)
```

```python
simple_options_df[['Amount','Price']] = simple_options_df[['Amount','Price']].fillna(0)
simple_options_df.sample(10)
```

```python
simple_options_df.isna().sum().sum()
```

```python
simple_options_df[['Price','Amount','Description']] = \
simple_options_df[['Price','Amount','Description']].applymap(lambda x: str(x).replace('$',''))
simple_options_df.sample(10)
```

```python
simple_options_df[['Price','Amount','Description']] = \
simple_options_df[['Price','Amount','Description']].applymap(lambda x: str(x).replace('(','-').replace(')','').replace(',',''))
simple_options_df['Price'] = simple_options_df['Price'].apply(lambda x: float(x)*100)
simple_options_df['Amount'] = simple_options_df['Amount'].apply(lambda x: float(x))
simple_options_df.info()
```

```python
simple_options_df.sample(10)
```

```python
simple_options_df['Expiration Date'] = simple_options_df['Description'].str.extract('(\d{1,2}/\d{1,2}/\d{2,4})')
simple_options_df['Strike'] = simple_options_df['Description'].str.extract('(\d+\.\d{2})')
simple_options_df['Strike'] = simple_options_df['Strike'].apply(lambda x: float(x))
```

```python
simple_options_df.head(1)
```

```python
simple_options_df.drop(columns = ['Description'],inplace=True)
```

```python
simple_options_df[['Activity Date', 'Expiration Date']] = \
simple_options_df[['Activity Date', 'Expiration Date']].apply(pd.to_datetime)
```

```python
simple_options_df.info()
```

```python
simple_options_df.sample(10)
```

```python
simple_options_df[simple_options_df['Quantity'].str.contains('S')].sample(3)
```

```python
# simple_options_df['Quantity'] = simple_options_df['Quantity'].apply(lambda x: int(x))
```

```python
simple_options_df['Quantity'] = simple_options_df['Quantity'].apply(lambda x: str(x).replace('S',''))
simple_options_df['Quantity'] = simple_options_df['Quantity'].astype(int)
```
